```python
def insta_scrapper(link,insta_sessionid):
    #form the link
    link=f'https://www.instagram.com/{link}/'
    import requests
    #sessionID unique for different instagrame ID
    SESSIONID =insta_sessionid
    #header which specific the machine which applys to the request
    headers = {"user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Mobile Safari/537.36 Edg/87.0.664.57",
           "cookie": f"sessionid={SESSIONID};"}
    #dynamic linktake account name name
    base_url = f'{link}?__a=1'
    print(base_url)
    #request the info from website
    response = requests.get(base_url,headers=headers)
    response = json.loads(response.text)
    #parse through nested json
    tmp_df=response['graphql']['user']['edge_owner_to_timeline_media']
    #convert json to DF
    df=pd.json_normalize(
        tmp_df,
        record_path=['edges']
    )
    #show names of columns
    #print(df.columns)
    #subset wanted info
    out_df=df[['node.owner.username','node.edge_liked_by.count',
               'node.edge_media_to_comment.count',
               'node.taken_at_timestamp']]
    #convert string to number
    out_df['Date']=pd.to_numeric(out_df['node.taken_at_timestamp'])
    #convert timestamp to date
    out_df['Date']=out_df['Date'].apply(lambda x: pd.datetime.fromtimestamp(x).date())
    #get earliest date boundary
    time_boundary=pd.to_datetime((date.today() - timedelta(days = 7)).strftime('%Y-%m-%d'))
    #subset based on time
    out_df=out_df.loc[(out_df['Date'] >= time_boundary)]
    #
    out_df=out_df.rename(columns={'node.owner.username':'influencer_name'})
    weekly_df=pd.DataFrame()
    naming={'Date':'post_count', 'node.edge_media_to_comment.count':'average_comment_count',
            'node.edge_liked_by.count':'average_like_count'
            }
    sum_df=out_df.groupby('influencer_name').agg({
        'Date':'size', 'node.edge_media_to_comment.count':'sum',
        'node.edge_liked_by.count':'mean'}).rename(columns=naming)
    sum_df['Date']=date.today().strftime('%Y-%m-%d')
    sum_df['follower_count']=int(response['graphql']['user']['edge_followed_by']['count'])
    sum_df['Date']=pd.to_datetime(sum_df['Date'])
    
    return sum_df
```

```python
def google_api():
    import time
    import random
    import datetime

    scope = ["https://spreadsheets.google.com/feeds",'https://www.googleapis.com/auth/spreadsheets',"https://www.googleapis.com/auth/drive.file","https://www.googleapis.com/auth/drive"]
    creds = ServiceAccountCredentials.from_json_keyfile_name("##############", scope)
    client = gspread.authorize(creds)
    spreadsheetName = "##############"
    sheetName = "##############"  # <--- please set the sheet name here.
    spreadsheet = client.open(spreadsheetName)
    sheet = spreadsheet.worksheet(sheetName)
    data = sheet.get_all_records()
    tmp_df=pd.DataFrame(data)
    influencer_df=tmp_df[tmp_df['Platform Used'].str.contains('Instagram')]['Handle/Name']
    influencer_df=influencer_df.replace('@','',regex=True).replace(' ','',regex=True)  
    influencer_df=influencer_df.str.split('//').str[0]
    df_list=pd.DataFrame()
    failed_list=[]
    for i in list(influencer_df):
        try:
            tmp=insta_scrapper(link=i,insta_sessionid='##############')
            time.sleep(random.uniform(1, 3)) 
            df_list=df_list.append(tmp)
            print(df_list.shape)
        except: 
            print('skip: ',i)
            failed_list.append(i)
            pass
    failed_df=pd.DataFrame(np.array(failed_list),columns=['fail_id'])
    failed_df['update_time']=datetime.date.today()

    return df_list.reset_index(),failed_df
  
```
